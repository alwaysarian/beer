# Beer Classification

For this project we will cluster the user from whom we will take review for three beer and use that information to suggest him beer.

```python
#For this project we will only use panda and numpy
import pandas as pd
import numpy as np
```

```python
#This is one time thing, where we will download that data from online
df = pd.read_csv("https://query.data.world/s/5wrkcltf0rqhyu2ahlayldrfg")
df.head()
```

```python
# For further reference we will be saving a copy in our system
# df.to_csv('out.csv')
```

```python
#since the data is too large we will trime the data and use only review_overall, review_profilename and beer_name
cleandata = df.loc[:,["review_overall","review_profilename", "beer_name"]]
cleandata.head()
```

```python
beer1 = "Budweiser"
overall1 = 1
beer2 = "Weihenstephaner Hefeweissbier"
overall2 = 5
beer3 = "Trappistes Rochefort 6"
overall3 = 3
```

```python
pd.options.mode.chained_assignment = None
```

```python
first = cleandata[(cleandata.beer_name == beer1)]

first['score'] = (first.review_overall - overall1)**2

first.head()
```

```python
second = cleandata[(cleandata.beer_name == beer2)]
second['score'] = (second.review_overall - overall2)**2

second.head()
```

```python
third = cleandata[(cleandata.beer_name == beer3)]
third['score'] = (third.review_overall - overall3)**2

third.head()
```

```python
firstdata = first.append(second)
final = firstdata.append(third)
final.head()
```

```python
grouppeddata = (final.groupby( [ "review_profilename"] ).count())
grp = grouppeddata[grouppeddata.review_overall == 3].reset_index()
grp.head()
```

```python
reviewer = grp.loc[:,"review_profilename"]
reviewer = reviewer.values.tolist()
reviewer
```

```python
finalresut = pd.DataFrame()
for allthename in reviewer:
    beername = allthename
    thefinalname = final[(final.review_profilename == beername)]
    finalresut = finalresut.append(thefinalname)
finalresut.shape
```

```python
grouppeddata1 = finalresut.groupby( [ "review_profilename"] ).sum() 
grouppeddata1 = grouppeddata1[grouppeddata1.score < 1].reset_index()
grouppeddata1
```
