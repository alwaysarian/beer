# Beer Classification - Using Overall Rating, Appearance, Taste and Aroma

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://query.data.world/s/5wrkcltf0rqhyu2ahlayldrfg")
df.head()
```

```python
cleandata = df.loc[:,["review_overall","review_aroma","review_appearance","review_profilename", "review_palate","review_taste",
                      "beer_name","beer_abv"]]
cleandata.head()
```

```python
beer1 = "Budweiser"
overall1 = 3
appearance1 = 3
aroma1 = 2
palate1 = 1
taste1 = 3
beer2 = "Weihenstephaner Hefeweissbier"
overall2 = 5
appearance2 = 5
aroma2 = 4
palate2 = 5
taste2 = 5
beer3 = "Trappistes Rochefort 6"
overall3 = 3
appearance3 = 4
aroma3 = 4
palate3 = 5
taste3 = 5
```

```python
first = cleandata[(cleandata.beer_name == beer1) & (cleandata.review_overall == overall1) & (cleandata.review_aroma == aroma1) 
                  & (cleandata.review_appearance == appearance1) & (cleandata.review_taste == taste1)]

first
```

```python
second = cleandata[(cleandata.beer_name == beer2) & (cleandata.review_overall == overall2) & (cleandata.review_aroma == aroma2) 
                   & (cleandata.review_appearance == appearance2) & (cleandata.review_taste == taste2)]
second
```

```python
third = cleandata[(cleandata.beer_name == beer3) & (cleandata.review_overall == overall3) & (cleandata.review_aroma == aroma3) 
                  & (cleandata.review_appearance == appearance3) & (cleandata.review_taste == taste3)]
third
```

```python
firstdata = first.append(second)
final = firstdata.append(third)
final
```

```python
reviewer = final.loc[:,"review_profilename"]
```

```python
reviewer = reviewer.values.tolist()
reviewer
```

```python
finalresut = pd.DataFrame()
for allthename in reviewer:
    beername = allthename
    thefinalname = cleandata[(cleandata.review_profilename == beername) & (cleandata.review_overall == 5) 
                             & (cleandata.review_aroma == 5) & (cleandata.review_appearance == 5) 
                             & (cleandata.review_taste == 5)]
    #print(thefinalname)
    finalresut = finalresut.append(thefinalname)
finalresut
```

```python
beernamefinal = finalresut.loc[:,"beer_name"]
```

```python
beernamefinal = beernamefinal.values.tolist()
beernamefinal
```
